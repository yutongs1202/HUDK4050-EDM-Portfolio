### 1. Criteria
#### 1.1 Alex's criteria

#### 1.2 Our group's definition

1. Safety

    We take the crime_2010 excel sheet from: https://ucr.fbi.gov/crime-in-the-u.s/2010/crime-in-the-u.s.-2010/tables/10tbl08.xls/view
    
    We created our own "Crime Rate Adjusted" based on the "Violent Crime", "Property Crime", and "Population". We weighted heavier on "Violent Crime" (70%) because it's more threatening than "Property Crime" (30%).
    
    The "Crime Rate adjusted" is generated following the formula: (D2*0.7+I2*0.3)/C2
2. Urban

    For urban criteria, we based mainly on Locale of institution.
    
    As Alex wants to live the big city life, we only take the schools that fulfills "City: Large (population of 250,000 or more)".
3. Diversity

    We define diversity by including student from different race and gender.
    
    We first filtered out the school that are open only for specific student population, including:
    
    a. "Historically Black College and University"
    
    b. "predominantly black institution"
    
    c. "Alaska Native Native Hawaiian serving institution"
    
    d. "tribal college and university"
    
    e. "Asian American Native American Pacific Islander-serving institution"
    
    f. "Hispanic-serving institution"
    
    g. "Native American non-tribal institution"
    
    h. "men-only college"
    
    i. "women-only college"
    
    We looked more closely on the race issue based on the "Total share of enrollment of undergraduate degree-seeking students" by race：
    
    a. "Total share of enrollment of undergraduate degree-seeking students who are white” should be less than 80%;
    
    b. "Total share of enrollment of undergraduate degree-seeking students who are black" should be greater than 3%;
    
    c. "Total share of enrollment of undergraduate degree-seeking students who are Hispanic" should be greater than 3%;
    
    d. "Total share of enrollment of undergraduate degree-seeking students who are Asian" should be greater than 3%.
4. Quality

    For the school quality, we take the school that are:
    
    a. currently operating,

    b. is not distance-education-only education,

    c. offers Bachelor's degree in Education,

    d. offers graduate degree,

    e. completion rate higher than 60%, and
    
    f. admission rate smaller than 50%.

#### 2.2 Crime Dataset

```python
# read crime data
ACA1Crimedata = pd.read_csv("/Users/Wyr/Desktop/HUDK 4050/ACA/ACA 1/ACA1_dataset/Crime_2010.csv")
# ACA1Crimedata
ACA1Crimedata
```

```python
data[['INSTNM','CITY']]
# show the cities of the current school list
```

```python
city = ['Long Beach', 'Los Angeles', 'San Diego', 'Washington', 'Tampa', 'Chicago', 'New Orleans', 'Baltimore', \
        'Boston', 'Minneapolis', 'New York', 'Bronxville Village', 'Cleveland', 'Tulsa', 'Pittsburgh', \
        'Philadelphia', 'Nashville', 'Houston', 'Austin']
# the list of cities which the schools are in
state = ['CALIFORNIA', 'CALIFORNIA', 'CALIFORNIA', 'DISTRICT OF COLUMBIA', 'FLORIDA', 'ILLINOIS', 'LOUISIANA', \
        'MARYLAND', 'MASSACHUSETTS', 'MINNESOTA', 'NEW YORK', 'NEW YORK', 'OHIO', 'OKLAHOMA', 'PENNSYLVANIA', \
        'PENNSYLVANIA', 'TENNESSEE', 'TEXAS', 'TEXAS']
# the list of states which the cities are in

safeCity = []
for x in range(19):

    ACA1City = (ACA1Crimedata['City'] == city[x]) & (ACA1Crimedata['State'] == state[x])
    if float(ACA1Crimedata[ACA1City]['Crime Rate Adjusted']) <= 0.02:
        safeCity.append(city[x]) 

# get the list of cities which the crime rates are below 0.02

safeCity[8] = 'Bronx'

# change 'Bronxville Village' to 'Bronx'

print(safeCity)
```

```python
# join the current school list with safeCity
C = pd.Series(safeCity, name='CITY')
finalList = data.merge(C, left_on='CITY', right_on='CITY', how='right')

# rank the remaining institutions based on admission rates
finalList['rank'] = finalList['ADM_RATE'].rank(method='min')

# print instution name, city, admission rate, and rank
finalList.sort_values(by=['rank'])[['INSTNM','CITY','ADM_RATE','rank']]
```

We got a 15 school list after filtering ranked based on admission rates.
